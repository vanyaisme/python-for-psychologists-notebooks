### Slicing
**SYNTAX/METHODS**
Slicing allows you to extract a portion of a list. The syntax is list[start:stop:step] .

```python
rt_ms = [320, 415, 290, 480, 350]  # reaction times (ms) for 5 trials

# Basic slicing
print(rt_ms[1:4])   # [415, 290, 480] (index 1 to 3)
print(rt_ms[:3])    # [320, 415, 290] (start to index 2)
print(rt_ms[2:])    # [290, 480, 350] (index 2 to end)

# Step and negative indices
print(rt_ms[::2])   # [320, 290, 350] (every 2nd element)
print(rt_ms[::-1])  # [350, 480, 290, 415, 320] (reversed)
print(rt_ms[-3:-1]) # [290, 480]

# Slice assignment
rt_ms[1:3] = [400, 300]
print(rt_ms)        # [320, 400, 300, 480, 350]
```

### Sorting
**SYNTAX/METHODS**
You can sort lists in-place using sort() or create a new sorted list using sorted() .

```python
nums = [4, 1, 7, 3]
nums.sort()           # In-place sort: nums is now [1, 3, 4, 7]
nums.sort(reverse=True) # [7, 4, 3, 1]

words = ["anxious", "calm", "neutral"]
new_list = sorted(words)  # Returns new sorted list: ['anxious', 'calm', 'neutral']

# Using the key parameter for custom sorting
words.sort(key=len)       # Sort by word length: ['calm', 'anxious', 'neutral']
words.sort(key=str.lower) # Case-insensitive alphabetical

# Sorting a list of tuples or dicts
students = [("Alice", 85), ("Bob", 92), ("Charlie", 78)]
students.sort(key=lambda x: x[1])  # Sort by the second item (grade)
```

### Removing Elements
**SYNTAX/METHODS**
There are several ways to remove items from a list, depending on whether you know the value or the index.
Use del to remove an element by its index — like erasing one entry from a spreadsheet row:

```python
scores = [72, 85, 91, 88]
del scores[2]   # removes 91 (index 2) → [72, 85, 88]
print(scores)
```

```python
conditions = ["control", "treatment_a", "treatment_b", "treatment_c", "treatment_d"]

conditions.remove("treatment_b")  # Removes first occurrence. Raises ValueError if missing.
item = conditions.pop()    # Removes and returns the last item ("treatment_d")
first = conditions.pop(0)  # Removes and returns item at index 0 ("control")

del conditions[0]         # Deletes item at index 0
scores = [72, 85, 91, 88, 79]
del scores[1:3]           # Deletes a slice (removes 85 and 91)

scores.clear()            # Empties the list completely: []
```

| Method | Action |
| --- | --- |
| remove(val) | Deletes by value |
| pop(idx) | Deletes by index and returns it |
| del lst[idx] | Deletes by index or slice |
| clear() | Removes all items |

### enumerate()
**SYNTAX/METHODS**
When you need both the index and the value while looping, enumerate() is the Pythonic way to do it.

```python
conditions = ["control", "low_stress", "high_stress"]

# The Pythonic way
for i, cond in enumerate(conditions):
    print(f"Index {i}: {cond}")

# Starting the count at 1 (useful for menus)
for num, cond in enumerate(conditions, start=1):
    print(f"{num}. {cond}")

# Compare with the older range(len()) approach:
for i in range(len(conditions)):
    print(f"Index {i}: {conditions[i]}")

# Practical: find all positions where reaction time equals 320 ms
rt_raw = [320, 290, 320, 415, 320]
positions = [i for i, val in enumerate(rt_raw) if val == 320]
```

Note: This uses list comprehension syntax — covered fully in S17. For now, read it as “collect indices where rt equals 320”.

### Membership & Aggregation
**SYNTAX/METHODS**
Python provides built-in operators and functions to quickly check contents or calculate statistics on lists.

```python
conditions = ["control", "treatment_a", "treatment_b"]

# Membership testing
if "control" in conditions:
    print("Control condition is included.")

if "treatment_c" not in conditions:
    print("Treatment C was not run.")

# Aggregation functions
grades = [85, 92, 78, 90]
print(min(grades))  # 78
print(max(grades))  # 92
print(sum(grades))  # 345
print(len(grades))  # 4

# Average calculation
average = sum(grades) / len(grades)

# any() and all()
bools = [True, False, True]
print(any(bools))   # True (at least one is True)
print(all(bools))   # False (not all are True)
```

### Unpacking & Copying
**ADVANCED**
Unpacking assigns list elements directly to separate variables in one line. Copying a list requires care: a shallow copy creates a new list with the same items, but if those items are themselves lists (nested), they are still shared. A deep copy duplicates everything — including nested lists — so nothing is shared between the original and the copy.

```python
# Basic unpacking — assign each element to a separate variable
low, mid, high = [72, 85, 91]  # participant scores

# Star unpacking — first gets index 0, rest gets everything else
p1, *others = [82, 91, 78, 65, 88]
print(p1)      # 82
print(others)  # [91, 78, 65, 88]

first, *middle, last = [82, 91, 78, 65, 88]

# Copying lists
scores = [82, 91, 78]
fake_copy = scores          # WARNING: Both point to the same list!

# Shallow copies (creates a new list object)
copy1 = scores.copy()
copy2 = scores[:]
copy3 = list(scores)

# Deep copy (needed if the list contains other lists)
import copy
groups = [[82, 91], [78, 65]]  # two participant groups
deep = copy.deepcopy(groups)
```

You already know lists (Section 14). Python has two more commonly-used built-in collection types: a tuple (an ordered sequence, like a list but unchangeable) and a set (an unordered collection with no duplicates — not a sequence). Dictionaries (key-value pairs) follow in Section 16.

### Python's Four Built-in Collections
**CONCEPT**
A few terms used in the table: Mutable means you can change the collection after creating it (add, remove, or replace items). Ordered means items keep their original position — index 0 is always the first item. Duplicates allowed means the same value can appear more than once.
| Type | Ordered? | Mutable? | Duplicates? | Best for |
| --- | --- | --- | --- | --- |
| list [] | Yes | Yes | Yes | Sequences you modify |
| tuple () | Yes | No | Yes | Fixed data, multiple return values |
| set {} | No | Yes | No | Unique items, fast membership test |
| dict {k:v} | Yes (3.7+) | Yes | Keys: No | Key-value mapping |

### Creating & Using Tuples
**TUPLES**
A tuple is like a list, but immutable — once created, you cannot change its items. Use tuples for data that should not change (coordinates, RGB colours, database rows) or to return multiple values from a function.

```python
# Creating tuples
point     = (3, 7)             # parentheses optional
rgb       = (255, 128, 0)       # orange colour
single    = (42,)              # single-element MUST have trailing comma
from_list = tuple([1, 2, 3])  # convert list to tuple

# Access by index (same as list)
print(rgb[0])   # 255
print(rgb[-1])  # 0

# Unpacking — the most important tuple feature
x, y = point
print(f"x={x}, y={y}")

r, g, b = rgb
print(f"R={r} G={g} B={b}")

# Extended unpacking with *rest
first, *rest = (1, 2, 3, 4, 5)
print(first)  # 1
print(rest)   # [2, 3, 4, 5]  (rest is a list)

# Pythonic variable swap
a, b = 10, 20
a, b = b, a
print(f"a={a}, b={b}")  # a=20, b=10

# Tuples as dict keys (unlike lists!)
grid = {(0,0): "origin", (1,0): "right", (0,1): "up"}
print(grid[(0,0)])  # origin

# Iteration with tuples
students = [("Alice", 85), ("Bob", 72), ("Carol", 91)]
for name, score in students:
    print(f"{name}: {score}")
```

### Creating & Using Sets
**SETS**
A set stores unique items with no guaranteed order. It's fast at membership testing and ideal for deduplication.

```python
# Creating sets
responses = {"agree", "neutral", "disagree"}   # survey response types
empty_set = set()  # NOT {} — that creates an empty dict!
from_list = set([1, 2, 2, 3, 3, 3])  # {1, 2, 3} — auto-deduplicates
print(from_list)

# Membership test — instant regardless of set size, much faster than list
print("agree" in responses)   # True
print("unsure" in responses)  # False

# Add / remove
responses.add("unsure")          # add a new response type
responses.discard("disagree")  # remove; no error if missing
print(responses)

# Deduplicate a list (fastest method)
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
unique  = sorted(set(numbers))  # sorted restores order
print("Unique:", unique)
```

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">SET OPERATIONS</span>
<span style="color: #6b7280; font-size: 0.85em; margin-left: 10px;">Union, Intersection, Difference</span>

```python
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

# Union: A OR B
print(A | B)        # {1,2,3,4,5,6,7}

# Intersection: BOTH A AND B
print(A & B)           # {3,4,5}

# Difference: in A but NOT in B
print(A - B)        # {1,2}

# Symmetric difference: A or B but not both
print(A ^ B)        # {1,2,6,7}

# Subset / superset
C = {3, 4}
print(C.issubset(A))    # True
print(A.issuperset(C))  # True
print(A.isdisjoint({8,9}))  # True — no overlap

# Real-world: students in both classes
class_a = {"Alice", "Bob", "Carol"}
class_b = {"Bob", "Dave", "Eve"}
print("Only in A:", class_a - class_b)
print("In both:",   class_a & class_b)
print("Combined:",  class_a | class_b)
```

### Immutable Sets as Dict Keys
**FROZENSET**
A frozenset is an immutable version of a set. Immutable means it cannot be changed after creation — you cannot add or remove items. Because it never changes, Python allows a frozenset to be used as a dictionary key (regular sets cannot). Psychology example: a frozenset of fixed condition labels (e.g., frozenset({"control", "treatment_a"}) ) could serve as a key to store results per condition combination.

```python
# frozenset: immutable, hashable — can be a dict key
perms_admin  = frozenset({"read", "write", "delete"})
perms_viewer = frozenset({"read"})

roles = {perms_admin: "admin", perms_viewer: "viewer"}
print(roles[perms_admin])   # admin

print(perms_viewer.issubset(perms_admin))  # True
print(perms_admin & perms_viewer)           # frozenset({'read'})
```

### What is a Dictionary?
**CONCEPT**
A dictionary stores data as key–value pairs . Keys must be unique and immutable (meaning they cannot be changed — strings, numbers, tuples). Values can be anything. Think of it like a real dictionary: you look up a word (key) to get its definition (value).
Use a dictionary when you need to look something up by name, not by position. If you catch yourself using parallel lists (one list of names, one list of ages) — that's a sign you need a dictionary.
| Feature | List | Dictionary |
| --- | --- | --- |
| Access by | Index (0, 1, 2…) | Key (any immutable value) |
| Order | Always ordered | Ordered (Python 3.7+) |
| Best for | Sequences of same-type items | Labelled, named data |
| Access speed | Must check each item one by one — slower as data grows | Jumps directly to the key — fast even with large data |

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">SYNTAX</span>
<span style="color: #6b7280; font-size: 0.85em; margin-left: 10px;">Creating Dictionaries</span>
</div>

```python
# Literal syntax: {key: value, ...}
student = {
    "name": "Alice",
    "age": 22,
    "grade": "A",
    "passed": True
}
print(student)

# Empty dictionary — add entries later
scores = {}
scores["Math"] = 87
scores["Science"] = 93
print(scores)

# dict() constructor — same as literal syntax, different style
participant_info = dict(id="P01", age=22, condition="control")
print(participant_info)

# Number keys: useful for Likert-scale labels
likert = {1: "Strongly Disagree", 3: "Neutral", 5: "Strongly Agree"}
print(likert[3])  # Neutral
```

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">CRUD</span>
<span style="color: #6b7280; font-size: 0.85em; margin-left: 10px;">Access, Update, Add & Delete</span>

```python
student = {"name": "Alice", "age": 22, "grade": "A"}

# READ — access by key (KeyError if missing!)
print(student["name"])          # Alice

# Safe read with .get() — returns None or a default instead of crashing
print(student.get("grade"))      # A
print(student.get("email"))      # None — key missing, no crash
print(student.get("email", "N/A")) # N/A — custom default

# UPDATE — reassign existing key
student["age"] = 23
print(student["age"])             # 23

# ADD — assign to a new key
student["email"] = "alice@example.com"
print(student)

# DELETE — remove a key-value pair
del student["grade"]
print(student)

# .pop() removes AND returns the value
email = student.pop("email")
print("Removed email:", email)
print(student)

# .pop() with default — avoids KeyError if key not present
missing = student.pop("phone", "not found")
print(missing)  # not found
```

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">METHODS</span>

```python
participant = {"id": "P03", "age": 30, "condition": "experimental"}

# .keys() — all keys as a view
print(list(participant.keys()))    # ['id', 'age', 'condition']

# .values() — all values as a view
print(list(participant.values()))  # ['P03', 30, 'experimental']

# .items() — key-value pairs as tuples
print(list(participant.items()))   # [('id','P03'), ('age',30), ...]

# Membership test — 'in' checks keys
print("id" in participant)       # True
print("email" in participant)      # False

# .update() — merge another dict in (overwrites existing keys)
participant.update({"age": 31, "session": 2})
print(participant)

# len() — number of key-value pairs
print(len(participant))  # 4

# .clear() — empties the dictionary
backup = participant.copy()   # .copy() creates an independent copy — changes to backup won't affect participant
participant.clear()
print(participant)   # {}
print(backup)   # still has data
```

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">ITERATION</span>
<span style="color: #6b7280; font-size: 0.85em; margin-left: 10px;">Looping over Dictionaries</span>

```python
scores = {"Math": 87, "Science": 93, "English": 79}

# Loop over keys (default)
for subject in scores:
    print(subject)  # Math, Science, English

# Loop over values
for score in scores.values():
    print(score)   # 87, 93, 79

# Loop over key-value pairs — most common pattern
for subject, score in scores.items():
    print(f"{subject}: {score}")

# Practical: find highest-scoring subject
best_subject = None
best_score   = 0
for subject, score in scores.items():
    if score > best_score:
        best_subject = subject
        best_score   = score
print(f"Best subject: {best_subject} ({best_score})")

# Total average
average = sum(scores.values()) / len(scores)
print(f"Average score: {average:.1f}")
```

### Nested participant database
**WORKED EXAMPLE**
A common psychology use-case: store each participant’s data as a nested dictionary, where the outer key is a participant ID and the inner dictionary holds their attributes.

```python
# Participant database: outer key = ID, value = dict of attributes
participants = {
    "P01": {"age": 21, "condition": "A"},
    "P02": {"age": 24, "condition": "B"},
}

# Access a top-level participant
print(participants["P01"])           # {'age': 21, 'condition': 'A'}

# Access a nested value: participant ID → attribute
print(participants["P01"]["age"])     # 21
print(participants["P02"]["condition"]) # B

# Loop: print each participant's details
for pid, data in participants.items():
    print(f"{pid}: age={data['age']}, condition={data['condition']}")
```

### Nested Dictionaries
**ADVANCED**
Dictionaries can contain other dictionaries as values. This is how you can store a dictionary inside another dictionary as a value.

```python
# A nested dictionary — each participant's data is itself a dictionary
students = {
    "alice": {"age": 22, "grade": "A", "scores": [90, 85, 92]},
    "bob":   {"age": 21, "grade": "B", "scores": [75, 80, 78]},
    "carol": {"age": 23, "grade": "A", "scores": [95, 88, 91]},
}

# Access nested value: outer key → inner key
print(students["alice"]["grade"])   # A
print(students["bob"]["scores"][0]) # 75 (first score)

# Loop: print each student's average score
for name, info in students.items():
    avg = sum(info["scores"]) / len(info["scores"])
    print(f"{name.capitalize()}: avg={avg:.1f}, grade={info['grade']}")

# Add data to a nested dict
students["alice"]["email"] = "alice@uni.nl"
print(students["alice"])
```

### Common Dictionary Mistakes
**PITFALLS**
| Mistake | What Happens | Fix |
| --- | --- | --- |
| d["missing"] | Python crashes with a key-not-found error | Use d.get("missing") |
| Mutable key: {[1,2]: "val"} | Python crashes — lists cannot be used as keys
keys | Use tuples: {(1,2): "val"} |
| Changing dict while looping | Python crashes with a dictionary-changed-during-loop error | Loop over list(d.items()) instead |
| Assuming insertion order (old Python) | Wrong order in Python < 3.7 | Use Python 3.7+ (insertion order guaranteed) |
| Shallow copy: d2 = d1 | Both variables point to same dict | Use d2 = d1.copy() or dict(d1) |

### What Are Comprehensions?
**CONCEPT**
A comprehension is a concise way to build a collection (list, dictionary, or set) by transforming and/or filtering a sequence of values — all in one expression (a piece of code that produces a value). They are common Python style : once mastered, your code becomes shorter, faster to read, and usually faster to execute.
Three types: list comprehensions [...] , dict comprehensions {k: v ...} , and set comprehensions {...} .
|  | Loop version | Comprehension version |
| --- | --- | --- |
| List of squares | result = [] for x in range(5): result.append(x**2) | [x**2 for x in range(5)] |

### One-line if/else for a single value
**TERNARY EXPRESSION**
A ternary expression (also called a conditional expression ) evaluates to one of two values depending on a condition — all in a single line. Syntax: value_if_true if condition else value_if_false .

```python
# Label a reaction time as "fast" or "slow" in one line
rt_ms = 285
result = "fast" if rt_ms < 300 else "slow"
print(result)  # "fast"

# Equivalent using a regular if/else block:
# if rt_ms < 300:
#     result = "fast"
# else:
#     result = "slow"
```

### List Comprehensions
**LIST**
Syntax: [expression for item in sequence] With filter: [expression for item in sequence if condition]

```python
# Step-by-step progression: loop → comprehension

# Traditional loop
squares_loop = []
for x in range(1, 6):
    squares_loop.append(x ** 2)
print(squares_loop)  # [1, 4, 9, 16, 25]

# Equivalent comprehension
squares = [x ** 2 for x in range(1, 6)]
print(squares)       # [1, 4, 9, 16, 25]

# With filter: only even numbers
evens = [x for x in range(20) if x % 2 == 0]
print(evens)

# Transform strings: strip whitespace and uppercase
raw = ["  alice ", " BOB", "carol  "]
clean = [name.strip().title() for name in raw]
print(clean)  # ['Alice', 'Bob', 'Carol']

# Filter list of dicts — students who passed
students = [
    {"name": "Alice", "score": 82},
    {"name": "Bob",   "score": 51},
    {"name": "Carol", "score": 91},
]
passed = [s["name"] for s in students if s["score"] >= 55]
print(passed)  # ['Alice', 'Carol']

# Conditional expression: label each score as Pass or Fail
scores = [72, 45, 88, 55, 91]
labels = ["Pass" if s >= 55 else "Fail" for s in scores]
print(labels)  # ['Pass', 'Fail', 'Pass', 'Pass', 'Pass']
```

### Dictionary Comprehensions
**DICT**
Syntax: {key_expr: value_expr for item in sequence if condition}

```python
# Square mapping
squares = {x: x**2 for x in range(1, 6)}
print(squares)   # {1:1, 2:4, 3:9, 4:16, 5:25}

# Build a gradebook using index — maps subject to score
subjects = ["Math", "Science", "English"]
scores   = [87, 93, 79]
gradebook = {subjects[i]: scores[i] for i in range(len(subjects))}
print(gradebook)

# Filter: only subjects with score >= 85
high_scores = {sub: s for sub, s in gradebook.items() if s >= 85}
print(high_scores)  # {'Math': 87, 'Science': 93}

# Invert a dictionary (swap keys ↔ values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)   # {1:'a', 2:'b', 3:'c'}

# Normalize string values: strip + lowercase
raw = {"Name": "  ALICE ", "City": " Amsterdam "}
clean = {k.lower(): v.strip().lower() for k, v in raw.items()}
print(clean)  # {'name': 'alice', 'city': 'amsterdam'}
```

### Set Comprehensions
**SET**
Syntax: {expression for item in sequence if condition} . Like a list comprehension but using {} and automatically deduplicating results.

```python
# All unique first letters from a word list
conditions = ["control", "control", "experimental", "experimental", "control"]
unique_conditions = {c for c in conditions}
print(unique_conditions)  # {'control', 'experimental'}

# Unique reaction times (removing duplicates)
rt_lengths = {len(str(c)) for c in conditions}
print(rt_lengths)  # unique string-lengths of each condition name

# Deduplicate scores while keeping only above-threshold values
scores = [3, -1, 3, 7, -2, 7, 5]
valid_scores = {s for s in scores if s > 0}
print(valid_scores)  # {3, 5, 7} — duplicates and negatives removed
```

### When to Use Each
**COMPARISON**
| You want to build… | Use |
| --- | --- |
| A list with optional filtering/transformation | [expr for x in iter if cond] |
| A dictionary from key-value pairs | {k: v for k, v in iter} |
| A unique set of values | {expr for x in iter} |
| Complex multi-step logic | Regular for loop (readability
wins) |

### Define once, call many times
**BASICS**
A function is a named, reusable block of code. You write it once and call it whenever you need it — no copying the same lines over and over. The keyword def tells Python you are defining a new function.
Parameter: the placeholder name in the definition (e.g. name below). Argument: the real value you pass in when calling the function (e.g. "Alice" ).

```python
def greet(name):             # name = parameter
    print(f"Hello, {name}!")

greet("P01")               # "P01" = argument
greet("P02")
```

### Avoid repeating the same logic — critical for PIP3
**REUSABLE CODE**
If you need to check whether a number is valid in several places, write that check once as a function and call it wherever needed. This avoids repeating the same logic and makes your code easier to fix — change it in one place and every call benefits.

```python
def validate_positive(value):
    """Returns the number if it is positive, otherwise returns None."""
    if value < 0:
        print("Error: cannot be negative")
        return None
    return value

speed1   = validate_positive(60)
speed2   = validate_positive(80)
distance = validate_positive(100)
if speed1 is None:
    print("Invalid input — using 0"); speed1 = 0
print(speed1, speed2, distance)
```

```python
# Keyword argument selects the calculation mode
def chase_time(speed1, speed2, distance, mode="towards"):
    if mode == "towards":
        time = distance / (speed1 + speed2)
        print(f"They meet after {time:.2f} hours")
    elif mode == "pursue":
        if speed1 == speed2:
            print("Same speed: will never catch up.")
        else:
            time = distance / abs(speed1 - speed2)
            print(f"Caught after {time:.2f} hours")

chase_time(60, 80, 100)                   # towards — default mode
chase_time(60, 80, 100, mode="pursue")   # pursue — override mode
```

### Local vs Global Scope
**SCOPE**
A local variable exists only inside the function where it was created — it disappears once the function finishes. A global variable is defined at the top level of your script, outside any function, and can be read from anywhere. Use the global keyword only when you must modify a global from inside a function (use sparingly — prefer return values instead).

```python
# Local scope: x inside the function
def double(n):
    result = n * 2  # 'result' is local to double()
    return result

x = double(5)
print(x)  # 10
# print(result)  # NameError: 'result' not defined here

# Global scope: counter exists outside the function
counter = 0

def increment():
    global counter  # without this, you'd get UnboundLocalError
    counter += 1

increment()
increment()
print(counter)  # 2

# ✅ Better: use return instead of global
def increment_val(n):
    return n + 1

counter = 0
counter = increment_val(counter)
counter = increment_val(counter)
print(counter)  # 2 (same result, no global)
```

### Returning Multiple Values & Unpacking
**MULTIPLE RETURNS**
Python functions can return multiple values by separating them with commas. Under the hood, Python packs them into a tuple , which you can unpack on assignment.

```python
# Return multiple values
def min_max(numbers):
    return min(numbers), max(numbers)  # returns a tuple

low, high = min_max([3, 1, 7, 2, 9])
print(f"Min: {low}, Max: {high}")  # Min: 1, Max: 9

# Ignore some return values with _ convention
_, high = min_max([3, 1, 7])
print(f"Only high: {high}")

# More complex: return stats as a dict (self-documenting)
def analyse(data):
    return {
        "count": len(data),
        "total": sum(data),
        "mean":  sum(data) / len(data),
        "min":   min(data),
        "max":   max(data),
    }

stats = analyse([72, 85, 91, 60, 78])
for key, val in stats.items():
    print(f"  {key}: {val}")
```

### map() — Transform Every Element
**SYNTAX**
The map(function, sequence) function applies a given function to every item in a sequence (such as a list) and returns the results. Wrap it in list() to see all values at once.
A lambda is a tiny, one-line function without a name: lambda x: x * 2 does the same job as def double(x): return x * 2 — it's just written in-line. Lambdas are useful with map() and filter() when the function is too simple to name separately.

```python
# Convert strings to integers
raw = ["450", "380", "510"]
rts = list(map(int, raw))
print(rts)  # [450, 380, 510]

# Apply a string method
conditions = ["control", "experimental", "control"]
upper_conds = list(map(str.upper, conditions))
print(upper_conds)  # ['CONTROL', 'EXPERIMENTAL', 'CONTROL']

# Using map with a lambda function
scores = [45, 72, 88]
scaled = list(map(lambda x: x / 100, scores))
print(scaled)  # [0.45, 0.72, 0.88]

# The same result using list comprehension — both approaches work
scaled_comp = [x / 100 for x in scores]
```

### filter() — Keep Matching Elements
**SYNTAX**
The filter(function, sequence) keeps only the items in a sequence for which the function returns True .

```python
rts = [450, 1200, 380, 950, 2100]

# Keep only fast response times
valid = list(filter(lambda x: x < 1000, rts))
print(valid)  # [450, 380, 950]

# Keep only filled condition labels
conditions = ["", "control", None, "experimental", ""]
clean = list(filter(None, conditions))
print(clean)  # ['control', 'experimental']

# filter(None, ...) removes falsy values (0, "", None, False, [])
entries = ["", "P01", None, "P02", ""]
clean_entries = list(filter(None, entries))
print(clean_entries)  # ['P01', 'P02']

# Compare with list comprehension
valid_comp = [x for x in rts if x < 1000]
```

### zip() — Pair Elements Together
**SYNTAX**
The zip(sequence1, sequence2, ...) function pairs up items from two or more sequences, position by position. It stops at the end of the shortest sequence.

```python
participants = ["P01", "P02", "P03"]
scores = [85, 92, 78]

# Pair elements together
paired = list(zip(participants, scores))
print(paired)  # [('P01', 85), ('P02', 92), ('P03', 78)]

# Iterate over multiple lists simultaneously
for pid, score in zip(participants, scores):
    print(f"{pid} scored {score}")

# Create a dictionary from two lists
score_dict = dict(zip(participants, scores))
print(score_dict)  # {'P01': 85, 'P02': 92, 'P03': 78}

# Unequal lengths: zip stops at the shortest list
print(list(zip(["P01", "P02", "P03"], [450, 380])))  # [('P01', 450), ('P02', 380)]
```

### Combining map, filter, zip
**PRACTICE**
You can combine these functions to create powerful data processing pipelines.

```python
participants = [
    {"id": "P01", "score": 85},
    {"id": "P02", "score": 42},
    {"id": "P03", "score": 91}
]

# Pipeline: filter passing participants, map to their IDs, zip with ranks
passing = filter(lambda p: p["score"] >= 50, participants)
names = map(lambda p: p["id"], passing)
ranks = [1, 2]

results = list(zip(ranks, names))
print(results)  # [(1, 'P01'), (2, 'P03')]
```

| --- | --- |
| map / filter | When you already have a named function (e.g., map(int, data) ) |
| Comprehensions | When you need a short calculation written inline. |

### Why Classes?
**CONCEPT**
In psychology research, you often need to track multiple pieces of data for each participant — their ID, age, condition, and reaction times. A class lets you bundle all that together.
Object-Oriented Programming (OOP) is a way of organising code around objects — bundles that combine data (attributes) and behaviour (methods). A class is the blueprint; an instance (object) is what you create from that blueprint.
Without classes, representing a participant requires parallel variables ( pid , age , condition ). With a class, one object holds all of them and knows what it can do with them.
|  | Procedural (no class) | OOP (with class) |
| --- | --- | --- |
| Data | Separate variables / dicts | Attributes on the object |
| Behaviour | Standalone functions | Methods bound to the object |
| Reuse | Copy-paste logic | Methods keep related behaviour bundled in one place |

<div style="margin: 20px 0 8px 0; padding: 8px 14px; background: #13162a; border-left: 3px solid #a78bfa; border-radius: 4px;">
<span style="color: #c4b5fd; font-weight: 600; font-size: 1.05em;">SYNTAX</span>
<span style="color: #6b7280; font-size: 0.85em; margin-left: 10px;">Defining a Class & Creating Instances</span>
</div>

```python
# Define a class
class Participant:
    """Represents a research participant."""

    # __init__ runs when you create an instance
    # 'self' refers to the new object being created
    def __init__(self, pid, age, condition):
        self.pid  = pid    # instance attribute
        self.age   = age
        self.condition = condition
        self.scores = []     # starts empty for each participant

    def greet(self):
        return f"Hi, I'm {self.pid}, age {self.age}, condition {self.condition}."

    def add_score(self, score):
        self.scores.append(score)

    def average(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

# Create instances (objects)
p01 = Participant("P01", 22, "experimental")
p02 = Participant("P02", 21, "control")

# Access attributes with dot notation
print(p01.pid)   # P01
print(p02.age)   # 21

# Call methods
print(p01.greet())

p01.add_score(85)
p01.add_score(92)
print(f"{p01.pid}'s average: {p01.average():.1f}")
```

### Class vs Instance Attributes
**CLASS VARIABLES**
Instance attributes (set via self. ) are unique per object. Class attributes are shared across all instances.

```python
class Student:
    school = "SICT"          # class attribute — shared by ALL students
    student_count = 0       # tracks how many students created

    def __init__(self, name, grade):
        self.name  = name    # instance attribute — unique per student
        self.grade = grade
        Student.student_count += 1  # update class attribute

    def info(self):
        return f"{self.name} ({self.grade}) @ {Student.school}"

s01 = Student("S01", "A")
s02 = Student("S02", "B")

print(s01.info())           # S01 (A) @ SICT
print(Student.student_count) # 2
print(s01.school)             # SICT (shared)
print(s02.school)             # SICT (same value)
```

### Exceptions — a quick recap
**CONCEPT**
You already know the basics of try/except earlier in this course. This section gives you the full toolkit : catching specific exception types, multiple except branches, the else and finally clauses, and raising your own errors.
| Situation | Without handling | With handling |
| --- | --- | --- |
| User types letters when a number is expected | Program crashes with ValueError | Show error message, ask again |
| File doesn’t exist | Crash with FileNotFoundError | Create file or inform user |
| Dividing by zero | Crash with ZeroDivisionError | Substitute a safe default value |

### Multiple except blocks
**SYNTAX**
One try block can have several except clauses — one per error type you want to handle differently. Python checks them top-to-bottom and runs the first matching one. Ordering matters: put the most specific exception before more general ones.

```python
raw = "0"   # try changing this to "hello" or ""

try:
    number = int(raw)
    result = 100 / number
except ValueError:
    print("That’s not a number. Please enter a digit.")
except ZeroDivisionError:
    print("Cannot divide by zero — try a number other than 0.")

print("Program continues.")
```

### try / except / else / finally
**FULL STRUCTURE**
Python's exception handling has four clauses. Each plays a distinct role:
| Clause | When it runs | Use for |
| --- | --- | --- |
| try | Always (first) | The risky code |
| except | Only if an error occurs in try | Handling the error |
| else | Only if NO error occurred in try | Code that should run on success |
| finally | Always, even after a return or error | Cleanup (close files, release resources) |

```python
filename = "rt_data.txt"

try:
    with open(filename, "r") as f:
        content = f.read()
except FileNotFoundError:
    print(f"File '{filename}' not found.")
    content = ""
except PermissionError as e:
    print(f"Permission denied: {e}")
    content = ""
else:
    print(f"File read successfully ({len(content)} chars).")
finally:
    print("Done — this always runs.")

print(f"Content preview: '{content[:30]}...'")
```

### Capturing the exception message (as e)
**SYNTAX**
When an exception is raised, Python creates an exception object that carries a message explaining what went wrong. Add as e after the exception type to capture that object in a variable. You can then print the message or log it for debugging.

```python
raw = "hello"

try:
    score = int(raw)
except ValueError as e:
    print("Error type:", type(e).__name__)   # ValueError
    print("Message:", str(e))               # invalid literal for int()...
```

type(e).__name__ gives you the exception class name as a string. str(e) gives you the human-readable message Python would normally print in the traceback.

### Raising exceptions (raise)
**SYNTAX**
You can raise an exception yourself with the raise keyword. This is useful when you detect bad input that Python wouldn’t catch automatically.

```python
def validate_rt(rt_ms):
    if rt_ms < 0:
        raise ValueError("Reaction time cannot be negative")
    print(f"Recorded reaction time: {rt_ms} ms")

try:
    validate_rt(-150)
except ValueError as e:
    print("Invalid input:", e)
```

A bare raise inside an except block re-raises the same exception — useful when you want to log the error but still let it propagate:

```python
try:
    score = int("bad")
except ValueError:
    print("Logging error and re-raising...")
    raise   # sends the exception up the call stack
```
