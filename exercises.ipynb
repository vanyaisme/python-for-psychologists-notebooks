```python
def toSmurf(aList):
    exchangesNeeded = True
    amountOfPasses = len(aList) - 1 # Fix 1: subtract 1
    while amountOfPasses > 0 and exchangesNeeded: # Fix 2: use bool directly
        exchangesNeeded = False
        for i in range(amountOfPasses):
            if aList[i] > aList[i + 1]:
                exchangesNeeded = True
                #swap
                temp = aList[i] # Fix 3: need temp variable
                aList[i] = aList[i + 1]
                aList[i + 1] = temp
        amountOfPasses = amountOfPasses - 1

aList = []
F = open("file.txt")
for line in F:
    aList.append(int(line.strip())) # Fix 4: Convert to int and strip
F.close()
toSmurf(aList)
print(aList) # Fix 5: Capitalization
```
