```python
import sys; sys.argv = ['script.py', '4', '8', '12', '16']  # ← edit these values to test
# ────────────────────────────────────────────────────────────
import sys

def validate(val_str):
    try:
        n = float(val_str)
    except ValueError:
        print("Error: not a number")
        return None
    if n < 0:
        print("Error: negative")
        return None
    return n

speed1   = validate(sys.argv[1])
speed2   = validate(sys.argv[2])
distance = validate(sys.argv[3])
mode     = sys.argv[4]

if None in (speed1, speed2, distance):
    print("Invalid input, exiting.")
    sys.exit(1)
elif mode == "towards":
    t = distance / (speed1 + speed2)
    print(f"Meet after {t:.2f} h, at {speed1 * t:.2f} km from start")
elif mode == "pursue":
    if speed1 == speed2:
        print("Same speed — will never catch up!")
    else:
        t = distance / abs(speed1 - speed2)
        print(f"Caught after {t:.2f} h")
```

```python
import sys
import os

# ── Mock: change these to test different inputs ──────────────────────────────
sys.argv = ['script.py', 'open.txt', '1', '2']  # ← edit: filename, ch_a, ch_b
# ─────────────────────────────────────────────────────────────────────────────

# Create sample EEG data file if it doesn't exist
if not os.path.exists('open.txt'):
    import random; random.seed(1)
    with open('open.txt', 'w') as _f:
        for _ in range(50):
            _f.write(' '.join(f'{random.uniform(0.1,1.5):.4f}' for _ in range(8)) + '\n')
    print("Created sample open.txt (8-channel EEG data)")

import csv
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 1. Read file
filename = sys.argv[1]          # "open.txt" or "closed.txt"
ch_a_idx = int(sys.argv[2]) - 1  # user enters 1-8, convert to 0-indexed
ch_b_idx = int(sys.argv[3]) - 1

eeg = [[] for _ in range(8)]
with open(filename, "r") as f:
    reader = csv.reader(f, delimiter=' ', skipinitialspace=True)
    for row in reader:
        for i, val in enumerate(row):
            eeg[i].append(float(val))

# 2. Plot both channels
plt.plot(eeg[ch_a_idx], label=f"Channel {ch_a_idx+1}")
plt.plot(eeg[ch_b_idx], label=f"Channel {ch_b_idx+1}")
plt.legend()
plt.show()

# 3. Compute and print correlation
r, p = pearsonr(eeg[ch_a_idx], eeg[ch_b_idx])
print(f"r = {r:.3f}, p = {p:.3f}")
```
